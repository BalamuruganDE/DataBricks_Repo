```python
# Bronze Layer: Ingest raw CSV files from landing_raw volume and write to Delta tables
from pyspark.sql import SparkSession

# Define file paths for each entity
base_path = "/Volumes/salesproject/default/sales/Landingraw/"
files = {
    "customers": f"{base_path}customers_raw.csv",
    "transactions": f"{base_path}transactions_raw.csv",
    "accounts": f"{base_path}accounts_raw.csv",
    "opportunities": f"{base_path}opportunities_raw.csv",
    "leads": f"{base_path}leads_raw.csv"
}

# Define target Delta table names
bronze_schema = "salesproject.bronze_layer"
tables = {
    "customers": f"{bronze_schema}.customers_raw",
    "transactions": f"{bronze_schema}.transactions_raw",
    "accounts": f"{bronze_schema}.accounts_raw",
    "opportunities": f"{bronze_schema}.opportunities_raw",
    "leads": f"{bronze_schema}.leads_raw"
}

# Ingest each CSV and write to Delta
for entity, file_path in files.items():
    df = spark.read.format("csv").option("header", True).option("inferSchema", True).load(file_path)
    df.write.format("delta").mode("overwrite").saveAsTable(tables[entity])
    print(f"Ingested {entity} to {tables[entity]}")

# Optionally, display a sample from each table
display(spark.table(tables["customers"]).limit(5))
```
